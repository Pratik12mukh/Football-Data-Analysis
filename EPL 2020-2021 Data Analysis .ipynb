```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns 
%matplotlib inline
```

```python
epl_df=pd.read_csv("EPL_20_21.csv") # load dataset 
epl_df.head(5)
```

```python
epl_df.info()
```

```python
epl_df.describe()# describe 
```

```python
epl_df.isnull().sum()# checking the null value present in the column 
```

```python
#creating two column with help of given column
epl_df['Minpermatch']=epl_df['Mins']/epl_df['Matches'].astype(int)
epl_df['Goalpermatch']=epl_df['Goals']/epl_df['Matches'].astype(float)
epl_df.head()
```

```python
#Total Goals 
Total_goals=epl_df['Goals'].sum()
Total_goals
```

```python
#penalty Goals
Total_penalty_goals=epl_df['Penalty_Goals'].sum()
Total_penalty_goals
```

```python
# Total number of penalties attempts
Total_penaltyattempts=epl_df['Penalty_Attempted'].sum()
Total_penaltyattempts
```

```python
#pie chart for penalties missed vs scored
plt.figure(figsize=(13,6))
Pl_not_scored=epl_df['Penalty_Attempted'].sum()-Total_penalty_goals
data=[Pl_not_scored,Total_penalty_goals]
labels= ['Penalties missed','penalties scored']
colors=sns.color_palette('Set2')
plt.pie(data,labels=labels,colors=colors,autopct='%.0f%%')
plt.show()
```

```python
#unique position
epl_df['Position'].unique()
```

```python
#total forward players
epl_df[epl_df['Position']=='FW']
```

```python
#players from different nations
np.size(epl_df['Nationality'].unique())
```

```python
#most player from which countries 
nationality=epl_df.groupby('Nationality').size().sort_values(ascending=False)
nationality.head(10).plot(kind='bar',figsize=(12,6),color=sns.color_palette("magma"))
```

```python
# clubs with maximum players in their squad 
epl_df['Club'].value_counts().nlargest(5).plot(kind = 'bar',color=sns.color_palette('viridis'))
```

```python
# clubs with least players in their squad 
epl_df['Club'].value_counts().nsmallest(5).plot(kind = 'bar',color=sns.color_palette('viridis'))
```

```python
#players based on age group 
under20 =epl_df[epl_df['Age']<=20]
age20_25=epl_df[(epl_df['Age']> 20) & (epl_df['Age']<= 25)]
age25_30=epl_df[(epl_df['Age']> 25) & (epl_df['Age']<= 30)]
above_30=epl_df[epl_df['Age']>30]
```

```python
x= np.array([under20['Name'].count(),age20_25['Name'].count(),age25_30['Name'].count(),above_30['Name'].count()])
mylabels=['<-20','>20 & <25','>25 & <30','>30']
plt.title('Total players with age group', fontsize=20)
plt.pie(x,labels=mylabels,autopct='%.1f%%')
plt.show()
```

```python
#Total under 20 players in each club 
player_under20=epl_df[epl_df['Age']<20]
player_under20['Club'].value_counts().plot(kind ='bar',color=sns.color_palette('cubehelix'))
```

```python
#Total under 20 players in ManU
player_under20[player_under20['Club']=='Manchester United']
```

```python
#Total under 20 players in chelsea
player_under20[player_under20['Club']=='Chelsea']
```

```python
#average age of players of club with box plot 
plt.figure(figsize=(12,6))
sns.boxplot(x='Club',y='Age',data=epl_df)
plt.xticks(rotation=90)
```

```python
#checking and verify average age of players of club with box plot 
num_player=epl_df.groupby('Club').size()
data= (epl_df.groupby('Club')['Age'].sum())/num_player
data.sort_values(ascending=False)
```

```python
#total assist from each club 
Assits_by_clubs=pd.DataFrame(epl_df.groupby('Club',as_index=False)['Assists'].sum())
sns.set_theme(style='whitegrid',color_codes=True)
ax=sns.barplot(x='Club',y='Assists',data=Assits_by_clubs.sort_values(by='Assists'),palette='Set2')
ax.set_xlabel('Club',fontsize=0)
ax.set_ylabel('Assists',fontsize=20)
plt.xticks(rotation=75)
plt.rcParams['figure.figsize']=(20,8)
plt.title('plot of clubs vs Total Assists',fontsize=20)
```

```python
#Top 10 Assists
top10assist=epl_df[['Name','Club','Assists','Matches']].nlargest(n=10,columns='Assists')
top10assist
```

```python
Goals_by_club=pd.DataFrame(epl_df.groupby('Club',as_index=False)['Goals'].sum())
sns.set_theme(style='whitegrid',color_codes=True)
ax=sns.barplot(x='Club',y='Goals',data=Goals_by_club.sort_values(by='Goals'),palette='rocket')
ax.set_xlabel('Club',fontsize=30)
ax.set_ylabel('Goals',fontsize=20)
plt.xticks(rotation=75)
plt.rcParams['figure.figsize']=(20,8)
plt.title('plot of clubs vs Total Goals',fontsize=20)
```

```python
#most goals by players
top_10_goals=epl_df[['Name','Club','Goals','Matches']].nlargest(n=10,columns='Goals')
top_10_goals
```

```python
#Goals per match
goalpermatch=epl_df[['Name','Goalpermatch','Matches','Goals']].nlargest(n=10,columns='Goalpermatch')
goalpermatch
```

```python
#pie chart  goals with assist and without assist 
plt.figure(figsize=(14,7))
assists=epl_df['Assists'].sum()
data=[Total_goals-assists,assists]
labels=['Goals without asssist','Goals with assists']
color=sns.color_palette('Set1')
plt.pie(data,labels=labels,colors=color,autopct='%.0f%%')
plt.show()
```

```python
#top 10 players with most yellow cards
epl_yellow=epl_df.sort_values(by='Yellow_Cards',ascending=False)[:10]
plt.figure(figsize=(20,6))
plt.title('Players with the most yellow cards')
c=sns.barplot(x=epl_yellow['Name'],y=epl_yellow['Yellow_Cards'],label='Players',color='yellow')
ax.set_ylabel('Numbers of Yellow Cards')
c.set_xticklabels(c.get_xticklabels(),rotation=45)
c
```
